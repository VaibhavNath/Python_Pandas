```python
person = {
    'first':['Vaibhav','Sam','John'],
    'last' : ['Nath','Doe','Doe'],
    'email' : ['Vaibhav.nath@email.com', 'Sam.Ham@email.com','John.Doe@email.com'],
    'prog_lang':['C','Python','Java']
}
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(person)
df
```

```python
filt= (df['last']=='Doe')  #here we got a series not a dataframe
print(filt)
```

```python
df[filt]      #here we got a dataframe
```

```python
df.loc[filt,['email','last']]     #gives the index value with email and last name column value
```

```python
filt= (df['last']=='Doe') & (df['first']=='John')    #use of and operator
```

```python
df[filt]
```

```python
filt= (df['last']=='Nath') | (df['first']=='John')       #use of or operator
```

```python
df[filt]
```

```python
filt= (df['last']=='Nath') | (df['first']=='John')        #use of not operator
```

```python
df[~filt]
```

```python
df        #original dataframe does not changes
```

```python
df.set_index('prog_lang',inplace=True)        #change index value of dataframe permanently
df
```
