```python
person = {
    'first':['Vaibhav','Sam','John','A'],
    'last' : ['Nath','Doe','Doe','Doe'],
    'email' : ['Vaibhav.nath@email.com', 'Sam.Ham@email.com','John.Doe@email.com','ADoe@gmail.com']
}
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(person)
df
```

```python
df=df.sort_values(by = 'last')    #sort the dataframe alphabetically by reference of specific column.
df
```

```python
df=df.sort_values(by = 'last',ascending=False)     #sort dataframe in descending order.
df
```

```python
df=df.sort_values(by = ['last','first'] , ascending=False)     #sort dataframe in descending order.
df
```

```python
df=df.sort_values(by = ['last','first'] , ascending=[False,True])  
df
```

```python
df.sort_index()   #sort the dataframe
```

```python
df['last'].sort_values()
```
